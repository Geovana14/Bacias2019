```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
L = 200.0E3 #metros
dx = 1000.0 #metros

x = np.arange(0,L,dx)
```

```python
#Criando vetor T de temperatura
t0 = 20 #graus C
q0 = 0.050 #W/m2
k = 2 #W/m/K
rho = 2800.0 #kg/m3
H = 9.6E-10 #W/kg



T = t0 + (q0*x)/k - ((rho*H)*(x**2))/(2*k)
```

```python
plt.figure(figsize=(10,6))
plt.plot(T, x/1000, 'g')
plt.ylim(L/1000,0)
plt.xlabel("Temperatura em ($^\circ$C)")
plt.ylabel("Profundidade (Km)")

```

```python
# 2º modelo não considerando H constante
# Criando novo vetor Temperatura T2

z = np.arange(0,L,dx)
qm = 0.025 #W/m2
hr = 10.0e3 #m
rhoH = 2.0E-6 #W/m2

T2 = t0 + (qm*z)/k + ((hr**2)*rhoH/k)*(1-np.exp(-z/hr))
```

```python
plt.figure(figsize=(10,6))
plt.plot(T2, z/1000)
plt.ylim(L/1000,0)
plt.xlabel("Temperatura em ($^\circ$C)")
plt.ylabel("Profundidade (km)")
           
```
