```python
#IDEIA: perfil de temperatura numa dorsal meso-oceanica

t = 0 #contador

kappa = 1.0E-6 #difusividade termica (m2/s)
k = 2.0 # condutividade térmica W/m/K

L = 100.0E3 #Tamanho do modelo (m)
dx = 1000.0 #Espacamento dos pontos (m)
z = np.arange(0.0,L,dx) #profundidade (m)

T0 = 0.0 #Temperatura inicial na superficie ºC
T1 = 1300.0 #Temperatura inicial no manto ºC

T = z*0+T1 #Para criar um vetor com o mesmo tamanho de z, mas todo zerado. Soma-se T1 para deixar com valores constantes = T1
T[0] = T0 #Mudar o primeiro valor do vetor de cima

plt.plot(T,z/1000,"k",label="0 Ma",lw='2')
plt.ylim(L/1000,-2)
plt.xlabel("Temperatura ($^\circ$C)")
plt.ylabel("Profundidade ($km$)")

tmax = 40.0E6 # anos totais de simulacao
dt = 1000.0 # passo de tempo entre iteracoes temporais
dt_seg = dt*365.0*24.0*60.0*60.0 # passo de tempo em segundos

t_print = 5.0E6 #Passo para printar curvas intermediárias - anos

q_acumulado = []
t_acumulado = []

while t<tmax:
    # Tfuturo = T[1:-1] + kappa*dt_seg*(T[2:] - 2*T[1:-1] + T[:-2])/dx**2 
    # T[2:]=vizinhos da frente (direita), T[1:-1]=indices do meio, T[:-2]=vizinhos de tras (esquerda)
    # T[1:-1] = Tfuturo # atualizo os valores centrais de T. T das extremidades sao fixos!
    
    #COMPACTANDO O MODELO
    T[1:-1] = T[1:-1] + kappa*dt_seg*(T[2:] - 2*T[1:-1] + T[:-2])/dx**2
    
    #Aqui vai imprimir em intervalos intermediarios entre tmax e t de 5 Ma.
    if t>t_print:
        t_print=t_print+5.0E6
        plt.plot(T,z/1000,"b",lw='1.5')
    
    q = k*(T[1]-T[0])/dx
    q_acumulado = np.append(q_acumulado,q)
    t_acumulado = np.append(t_acumulado,t)
    
    t += dt
    
plt.plot(T,z/1000,"r",label="+ 40 Ma",lw='2') #imprime só quando chega em 40 Ma
plt.legend(loc='lower left')
plt.title("GRADIENTE TÉRMICO DAS DORSAIS")
```

O python não pega o termo final quando x[2:4], por exemplo. Ele vai pegar o 2º e o 3º termo, mas o 4º não.
Para ele pegar o vetor até o final, posso escrever x[2:]. Se quiser pegar do começo, escrevo [:4].

O gráfico mostra diferentes curvas para a taxa térmica de uma listosfera oceânica. Depois de 40 Ma, a litosfera oceânica será mais espessa e mais distante da dorsal (o modelo também considera variação na distância dx).

```python
plt.plot(t_acumulado/1.0E6,q_acumulado)
plt.xlabel("Tempo em Ma")
plt.ylabel("Fluxo térmico (W/m$^2$)")
plt.ylim(0.,0.2)
plt.title("FLUXO TÉRMICO DA LITOSFERA OCEÂNICA (EM SUPERFÍCIE)")
```

O gráfico acima mostra que tem um fluxo térmico alto nas dorsais e decai à medida que se afasta das dorsais
