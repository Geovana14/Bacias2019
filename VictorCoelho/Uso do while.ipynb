```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import numpy as np
#import matplotlib.pyplot as plt
t = 0
kappa= 1.0e-6 #m2/s difusividade termica
k = 2.0 #W/m/Kelvincondutividade termica

L = 100.0E3 #m - tamanho do meu problema - profundidade
dx = 1000.0 #m - espacamento entre os pontos em profundidade

T0 = 0.0 #graus C (temepratura na superficie)
T1 = 1300.0 #graus C (temperatura no manto)

z = np.arange(0,L,dx)
T = z*0+T1 # comando para criar um vetor T com mesmo tamanho de z com valores cte =s a T1
T[0] = T0

plt.plot(T,z/1000, "k")
plt.ylim(L/1000,-2)

tmax = 40.0E6 #anos totais de simulacao
dt = 1000.0 #passo de tempo entre iteracoes temporais
dt_seg = dt *(365*24*60*60) #passo de tempo em segundos

q_acumulado = []
t_acumulado = []

t_print=5.0E6

while t<tmax:
    #Tfuturo = T[1:-1]+ kappa*dt_seg*(T[2:]-2*T[1:-1]+T[:-2])/dx**2 --- posso resumir essas duas linhas na linha de baixo
    #T[1:-1] = Tfuturo
    T[1:-1] = T[1:-1]+ kappa*dt_seg*(T[2:]-2*T[1:-1]+T[:-2])/dx**2
    if t>t_print:
        t_print=t_print+5.0E6
        plt.plot(T,z/1000,"b")
    
    q = k*(T[1]-T[0])/dx
    
    q_acumulado = np.append(q_acumulado,q)
    t_acumulado = np.append(t_acumulado/1.0E6,t)
    t= t+dt
    
plt.plot(T,z/1000, "r")
plt.ylabel("Profundidade em (km)")
#plt.plot(t_acumulado/1.0E6,q_acumulado,)
#plt.ylabel ("fluxo termico em (W/m$^2$)")
#plt.xlabel ("tempo em milhoes de anos")
#plt.ylim([0,0.2])    
```

```python
plt.plot(t_acumulado/1.0E6,q_acumulado,)
plt.ylabel ("fluxo termico em (W/m$^2$)")
plt.xlabel ("tempo em milhoes de anos")
plt.ylim([0,0.2])    
```
