```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kappa = 1.0E-6
k = 2.0

L = 100.0E3
dx = 1.0E3

T0 = 0.0
T1 = 1300.0

x = np.linspace(0,L,int(L/dx)+1)
T = np.zeros(x.size) + T1
T[0] = T0

# plt.figure(figsize=(5,10))
plt.plot(T,x/1.0E3,"k")
plt.ylim(L/1.0E3,-2)

t = 0
t_max = 40.0E6
dt = 1.0E3
dt_seg = dt*365.25*24.*60.*60.
t_print = 5.0E6

q_acumulado = []
t_acumulado = []

while t<=t_max:
    Tfuturo = T[1:-1] + kappa * dt_seg * (T[2:]-2*T[1:-1]+T[:-2]) / dx**2
    T[1:-1] = Tfuturo
    
    if t>t_print:
        t_print = t_print + 5.0E6
        plt.plot(T,x/1.0E3,"b")
        
    q = k * (T[1]-T[0]) / dx
    q_acumulado = np.append(q_acumulado,q)
    t_acumulado = np.append(t_acumulado,t)
    t += dt
plt.plot(T,x/1.0E3,"r")
plt.xlabel("Temperatura ($^\circ$C)")
plt.ylabel("Profundidade (km)")
```

```python
plt.plot(t_acumulado/1.0E6,q_acumulado*1.0E3)
plt.xlabel("Tempo (Ma)")
plt.ylabel("Fluxo Térmico (mW/m$^2$)")
plt.ylim(0,200)
```
