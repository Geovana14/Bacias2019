```python
import numpy as np
import matplotlib.pyplot as plt
```

# Caso com H constante

```python
L = 200.0E3 # tamanho do modelo [m]
dx = 1000.
x = np.linspace(0,L,int(L/dx)+1)
T0 = 20. 
k = 2.
rho = 2700. # [kg/m3]

q0 = 0.05
H = 9.6E-10

T = T0 + (q0*x/k) - (rho*H*x*x/(2*k))
```

```python
plt.plot(T,x/1.0E3)
plt.ylim(L/1.0E3,0)
plt.xlabel("Temperatura ($^\circ$C)")
plt.ylabel("Profundidade de (km)")
```

# Caso com H variável

```python
qm = q0 / 2.
hr = 10.0E3
H0 = H
rhoH = 2.0E-6
T = T0 + (qm*x/k) + (hr*hr*rhoH/k)*(1-np.exp(-x/hr))
```

```python
plt.plot(T,x/1.0E3)
plt.ylim(L/1.0E3,0)
plt.xlabel("Temperatura ($^\circ$C)")
plt.ylabel("Profundidade de (km)")
```
