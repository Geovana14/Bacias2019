```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 200.0E3 #m
dx = 1000.0 #m
x = np.arange(0,L,dx)
```

```python
T0 = 20.0 #graus C
k = 2 #W/m/K
q0 = 0.050 #W/m2
rho = 2800.0 #kg/m3
H = 9.6E-10 #W/kg


T = T0 + q0*x/k - (rho*H)*x**2/(2*k)
```

```python
plt.plot(T,x/1000)
plt.ylim(L/1000,0)
plt.xlabel("Temperatura ($^\circ$C)")
plt.ylabel ("Profundidade (km)")
```

```python
z = np.arange(0,L,dx)
qm = 0.025 #W/m2
rhoH = 2.0E-6 #W/m3
hr = 10.0E3 #m

T = T0 + qm*z/k + (hr**2*rhoH/k)*(1.0-np.exp(-z/hr))

plt.plot(T,z/1000)
plt.ylim(L/1000,0)
plt.xlabel("Temperatura ($^\circ$C)")
plt.ylabel ("Profundidade (km)")
```
