```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 0

kappa = 1.0E-6 #m2/s - difusividade térmica

k = 2.0 #W/m/K - condutividade térmica

L = 100.0E3 #m - tamanho do problema
dx = 1000.0 #m - espaçamento entre os pontos

T0 = 0 #graus C - temperatura na superfície
T1 = 1330 #graus C - temperatura inicial no manto

z = np.arange(0,L,dx)
T = z*0+T1 #vetor com o mesmo tamanho de z, mas com valor constante
T[0] = T0

plt.plot(T,z/1000,"k")
plt.ylim(L/1000,-2)

tmax = 40.0E6 #anos totais de simulação
dt = 1000.0 #passo de tempo entre interações temporais
dt_seg = dt*365.*24.*60.*60. #passo de tempo em segundos

q_acumulado = []
t_acumulado = []

t_print = 5.0E6

while t<tmax:
    #Tfuturo = T[1:-1]+kappa*dt_seg*(T[2:]-2*T[1:-1]+T[:-2])/dx**2
    #T[1:-1] = Tfuturo
    
    T[1:-1] = T[1:-1]+kappa*dt_seg*(T[2:]-2*T[1:-1]+T[:-2])/dx**2
    
    if t>t_print:
        t_print = t_print+5.0E6
        plt.plot(T,z/1000,"b")
    
    q = k*(T[1]-T[0])/dx
    q_acumulado = np.append(q_acumulado,q)
    t_acumulado = np.append(t_acumulado,t)
    
    t = t+dt
    
plt.plot(T,z/1000,"r")
plt.ylabel ("Profundidade (km)")
```

```python
plt.plot(t_acumulado/1.0E6,q_acumulado)
plt.xlabel ("Tempo em Ma")
plt.ylabel ("Fluxo térmico (W/m$^2$)")
plt.ylim([0,0.2])
```
