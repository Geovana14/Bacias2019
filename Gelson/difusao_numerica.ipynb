```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 0
kappa = 1.0E-6 # m2/s --> difusidade térmica
k = 2.0 # W/(mK) - condutividade
L = 100.0E3 # tamanho do meu problema
dx = 1000.0 # espaçamento entre os pontos
z = np.arange(0, L, dx)
T0 = 0 # graus C --> temperatura na superficie
T1 = 1300.0 # Graus C --> temperatura inicial do manto
T = z*0+T1 #cria um vetor to tamanho z com valor constante de T1
T[0] = T0
q_acumulado = []
t_acumulado = []

tmax = 40.0E6 # Anos totais de simulação
dt = 1000.0 # Passo de tempo entre iterações temporais
dt_seg = dt * 365.*24.*60.*60. # Passo de tempo em segundos

plt.plot(T, z/1000, 'k')
plt.ylim(L/1000, -2)

t_print = 5.0E6
while t<tmax:
    #Tfuturo = T[1:-1] + (kappa*dt_seg*(T[2:]-2*T[1:-1]+T[:-2])/dx**2)
    #T[1:-1] = Tfuturo
    
    T[1:-1] = T[1:-1] + (kappa*dt_seg*(T[2:]-2*T[1:-1]+T[:-2])/dx**2)
    
    if t>t_print: 
        t_print=t_print+5.0E6
        plt.plot(T, z/1000, 'b')
        
    q = k*(T[1]- T[0])/dx
    q_acumulado = np.append(q_acumulado, q)
    t_acumulado = np.append(t_acumulado, t)
    
    t = t + dt
    
plt.plot(T, z/1000, 'r')

plt.ylabel('Profundidade (Km)')
plt.xlabel('Temperatura (C)')
plt.ylim(L/1000, -2)
```

```python
plt.plot(t_acumulado/1.0E6, q_acumulado)
plt.xlabel("Tempo em Ma")
plt.ylabel('Fluxo térmico')
plt.ylim(0, 0.2)
```
