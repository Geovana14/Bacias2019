```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 200.0E3 # m
dx = 1800.0 # m
x = np.arange(0, L, dx)
```

```python
T0 = 20.0 # graus celsius 
k = 2 # W k-1 m-1
q0 = 0.050 # W/m2
rho = 2800.0 # kg/m3
H = 9.6E-10 # W/kg

T = T0 + (q0/k)*x - ((rho*H)/(2*k))*(x**2)
```

```python
plt.plot(T, x/1000)
plt.ylim(L/1000, 0) 
plt.ylabel("Profundidade (m)")
plt.xlabel("Temperatura ($^\circ$C)")
```

```python
z = np.arange(0, L, dx)
qm = 0.025 # W/m2
rhoH = 2.0E-6 # W/m3
hr = 10.0E3 # m

T = T0 + qm*z/k + (hr**2*rhoH/k)*(1.0 - np.exp(-z/hr))
```

```python
plt.plot(T, z/1000)
plt.ylim(L/1000, 0) 
plt.ylabel("Profundidade (m)")
plt.xlabel("Temperatura ($^\circ$C)")
```
