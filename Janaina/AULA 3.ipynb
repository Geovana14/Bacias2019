## AULA 3 - ESTRUTURA TÉRMICA

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 200.0E3 #200.000 m #espessura 
dx = 1000.0 #m
X = np.arange(0,L,dx) #posição espacial
```

```python
T0 = 20.0 #graus C - temperatura na superfície
k = 2 #W/m/K - condutividade térmica
q0 = 0.050 #W/m2 - fluxo de calor na superfície
rho = 2800.0 #kg/m3 - densidade na crosta
H = 9.6E-10 #W/kg

#EQUAÇÃO DE DIFUSÃO LEVANDO EM CONTA A PRODUÇÃO DE CALOR RADIOGÊNICO NO MODELO TODO
T = T0 + q0*X/k -(rho*H)*(X**2)/(2*k) 

#ESTAMOS ASSUMINDO QUE A PRODUÇÃO DE CALOR RAGIODÊNICO VARIA EM PROFUNDIDADE
#ASSUMIMOS QUE H ERA UMA CONSTANTE
```

```python
plt.plot(T,X/1000)
plt.ylim(L/1000,0)
plt.xlabel("Temperatura($\circ$C)")
plt.ylabel("Profundidade(km)")
plt.show()
```

```python
z =np.arange(0,L,dx)
qm = 0.025#W/m2 -fluxo de calor que vem só do manto
rhoH =2.0E-6 #W/m3 - 
hr = 10.0E3 #km - produção de calor radiogênico variando em profundidade

#EQUAÇÃO DE DIFUSÃO LEVANDO EM CONTA A PRODUÇÃO DE CALOR RADIOGÊNICO EM QUE A TEMPERATURA VARIA COM A PROFUNDIDADE
T = T0 + qm*z/k + (hr**2*(rhoH)/(k)) *(1.0-np.exp(-z/hr))

#ASSUMIMOS QUE H NÃO É CONSTANTE
```

```python
plt.plot(T,z/1000)
plt.ylim(L/1000,0)
plt.xlabel("Temperatura($\circ$C)")
plt.ylabel("Profundidade(km)")
plt.show()

#OBSERVE QUE INICIALMENTE TEMOS UMA CURVINHA EM Z=0 E DEPOIS TORNA-SE UMA RETA
#PARA OS PRIMEIROS 100 KM É BEM RAZOÁVEL ESTA CURVA

#A ASTENOSFERA É MECANICAMENTE INSTÁVEL

#A EQUAÇÃO QUE DEDUZIMOS É VÁLIDA SOMENTE PARA LITOSFERA
```
