### AULA 4 - DIFUSÃO NUMÉRICA

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
#DIFERENÇAS FINITAS

#Evolução do u ao longo do tempo

t=0
t_max=40.0E6 #anos totais de simulação

dt= 1000.0 #passo de tempo entre interações temporais
dt_seg=dt*365.*24.*60.*60. #passo de tempo em segundos

t_print=5.0E6

L=100.0E3 #m - tamanho do nosso modelo
dx=1000.0 #m - espaçamento entre os pontos

kappa=1.0E-6 #m²/s - difusividade térmica

z=np.arange(0,L,dx) #m - profundidade 

T0=0.0 #graus C - temperatura na superfície
T1=1300.0 #graus C - temperatura inicial no manto

T=z*0+T1 #vetor com o mesmo tamanho de z, mas com valores constantes iguais a T1
T[0]=T0

plt.plot(T,z/1000,"k")
plt.ylim(L/1000,-2)

k= 2.0 #W/m/K - condutividade térmica
q_acumulado = []
t_acumulado = []

while t<t_max:
    
    Tfuturo = T[1:-1] + kappa * dt_seg * (T[2:] - 2* T[1:-1] + T[:-2])/dx**2
    
    T[1:-1] = Tfuturo
    
    if t>t_print:
        t_print=t_print+5.0E6 #some mais 5 milhões de anos
        plt.plot(T,z/1000,"b")
        
    #FLUXO DE CALOR NA SUPERFÍCIE POR DIFERENÇAS FINITAS
        
    q = k*(T[1]-T[0])/dx
    
    q_acumulado = np.append(q_acumulado,q)
    t_acumulado = np.append(t_acumulado,t)
    
    t = t + dt

plt.plot(T,z/1000,"r")
plt.ylim(L/1000,-2)
plt.xlabel("Tempo")
plt.ylabel("Profundidade(km)")
```

ANÁLISE DO GRÁFICO

A temperatura inicial é dada pela curva preta. Nós temos uma camada de rocha a 1300ºC em contato com as rochas ao redor bem frias. 

Observe que a temperatura cai abruptamente nos primeiros milhões de anos e depois suaviza.

Cada curva representa a temperatura térmica em relação ao tempo.

Essa curva vermelha está a certa distância da dorsal que depende da taxa de espalhamento.

A litosfera ocênica espessa incialmente e depois estabiliza. As convecções abaixo reciclam continuamente o material, mantendo a temperatura da astenosfera constante e estabilizando a espessura da litosfera oceânica.

```python
plt.plot(t_acumulado/1.0E6,q_acumulado)
plt.xlabel("Tempo em milhoes de anos")
plt.ylabel("Fluxo Termica(W/m$^2$)")
plt.ylim([0, 0.2])

#O fluxo fica cada vez menor a medida que nos distanciamos da dorsal
```
