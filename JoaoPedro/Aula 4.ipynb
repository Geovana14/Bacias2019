```python
t=0
kappa=1.0E-6 #m2/s difusividade térmica

k=2.0

L=100.0E3 #m - tamanho do problema
dx=1000.0 #m - espaçamento entre pontos
T0=0.0 #graus C (temperatura na superfície)
T1=1300.0 #graus C (temperatura inicial do manto)
z=np.arange(0,L,dx)
T=z*0+T1 #vetor com o mesmo tamanho0 de z mas com valores constantes iguais a T1
T[0]=T0

plt.plot(T,z/1000,"k")
plt.ylim(L/1000,-2) #-2 é só pro 0 não ficar grudado na curva

tmax = 40.0E6 #anos totais de simulação
dt = 1000.0 #passo de tempo entre iterações temporais
dt_seg=dt*365.*24.*60.*60. #passo de tempo em segundos

q_acumulado=[]
t_acumulado=[]


t_print=5.0E6

while t<tmax:
    Tfuturo= T[1:-1] + kappa*dt_seg*(T[2:]-2*T[1:-1]+T[:-2])/dx**2
    
    T[1:-1]=Tfuturo
    
    if t>t_print:
        t_print=t_print+5.0E6
        plt.plot(T,z/1000,"b")
    
    q=k*(T[1]-T[0])/dx
    
    
    q_acumulado=np.append(q_acumulado,q)
    t_acumulado=np.append(t_acumulado,t)
    
    t=t+dt

plt.xlabel("Temperatura ºC")
plt.ylabel("Profundidade (km)")
    
plt.plot(T,z/1000,"r")


```

o grafico acima mostra a variação de temp como se estivessemos presos na crosta que esta se distanciando da dorsal e ficando mais velha. a base litosférica oceãnica tem temperatura dinamicamente mantida pelas correntes de convecção que vão reciclando esta base. No nosso modelo isso se mantém devido as condições de contorno impostas. o L do modelo corresponde à profundidade e não a distâmcia da dorsal.

podemos calcular o fluxo de calor tb por aproximações com diferenças finitas.

```python
plt.plot(t_acumulado/1.0E6,q_acumulado)
plt.xlabel("Tempo em Ma")
plt.ylabel("Fluxo térmico (W/m$^2$))
```
