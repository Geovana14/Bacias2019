```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 0
t_max = 40.0E6 #anos - Tempo maximo
dt = 1000.0 #anos - Passo de tempo em anos
dt_seg = dt*365.*24.*60.*60. # s - Tempo de cada passo 
kappa = 1.0E-6 #m2/s - difusividade termica
k = 2.0 # W/m/K - Condutividade termica
L = 100.0E3 #m - tamanho modelo
dx = 1.0E3 #m - espacamento entre pontos
z = np.arange(0,L,dx)

T0 = 0.0 #graus C - temperatura superficie 
T1= 1300.0 #graus C - temperatura inicial no manto
T = z*0 +T1
T[0] = T0

plt.plot(T,z/1000,'k')
plt.ylim(L/1000,-2)

t_print= 5.0E6

q_acumulado = []
t_acumulado = []
while t < t_max:
    #T_fut= T[1:-1] + kappa*dt_seg*(T[2:] - 2*T[1:-1] + T[:-2])/dx**2
    #T[1:-1] = T_fut
    
    T[1:-1] = T[1:-1] + kappa*dt_seg*(T[2:] - 2*T[1:-1] + T[:-2])/dx**2
    
    #if (t%t_print==0 and t >0.):
    if t>t_print:
        t_print=t_print+5.0E6
        plt.plot(T,z/1000,'b')
    
    q = k*(T[1]-T[0])/dx
    q_acumulado = np.append(q_acumulado,q)
    t_acumulado = np.append(t_acumulado,t)
    
    t = t + dt
    
plt.plot(T,z/1000,'r')
```

```python
plt.plot(t_acumulado/1.0E6,q_acumulado)
plt.xlabel("Tempo em Ma")
plt.ylabel("Fluxo Termico (W/m$^2$)")
plt.ylim(0,0.2)
```
